```python
import numpy as np
import pandas as pd

# Введені дані (всі оцінки в єдиному списку)
data = [
    52, 54, 70, 73, 75, 52, 82, 93, 95, 82,
    60, 62, 50, 79, 92, 91, 81, 53, 80, 70,
    52, 82, 93, 95, 82, 55, 60, 62, 50, 90,
    52, 54, 70, 73, 75, 52, 82, 93, 95, 82,
    82, 82, 93, 55, 60, 62, 50, 79, 28, 100,
    70, 73, 75, 52, 58, 73, 75, 52, 82, 84,
    52, 54, 70, 73, 75, 52, 82, 93, 95, 82,
    100, 66, 84, 93, 95, 82, 82, 93, 95, 82,
    84, 60, 62, 50, 79, 52, 54, 70, 73, 73,
    73, 75, 52, 82, 66, 70, 73, 75, 52, 54
]

# Упорядкування вибірки
sorted_data = sorted(data)

# Виведення впорядкованої вибірки
sorted_data
```

```python
# Розрахунок частот
frequency_table = pd.Series(sorted_data).value_counts().sort_index()

# Перетворення в DataFrame для зручності
frequency_df = pd.DataFrame({
    "Оцінка": frequency_table.index,
    "Частота": frequency_table.values
})

frequency_df
```

```python
# @title Частота

from matplotlib import pyplot as plt
frequency_df['Частота'].plot(kind='hist', bins=20, title='Частота')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
